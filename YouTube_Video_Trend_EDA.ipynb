# **5.Convert datatype of column published_at from object to pandas datetime.**

```python

df1['published_at'] = pd.to_datetime(df1['published_at'])
df1['published_at']
```

# **6.Create a new column as 'published_month' using the column published_at (display the months only)**

```python
df1['published_month'] = df1['published_at'].dt.month
df1['published_month']
```

# **7.Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'**

```python

d = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
df1['published_month'] = df1['published_month'].replace(d)
df1['published_month']
```

# **8.the number of videos published each month and arrange the months in a decreasing order based on the video count**

```python

df1.groupby(by='published_month')['video_id'].count().sort_values(ascending=False)
```
